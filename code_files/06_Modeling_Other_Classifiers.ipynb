## Problem Statement

### Given a dataset with demographic and borrowing history data for accounts classified as defaulting or not defaulting in October-2005, can I build a supervised model that performs better than identifying only members of the negative non-default class (baseline model) while minimizing the misclassification of either class?  In the context of credit-card lending, if I can predict accounts as belonging to the defaulting class, I want to minimize the number of predicted defaulters who did not actually default that October (lost revenues) while minimizing the number of predicted non-defaulters who did end up defaulting (lost profits).

## C. Logistic-Regression Modeling:

### We apply a GridSearch routine below to select hyperparameters that optimize accuracy as a first step.

```python
lr = LogisticRegression()

pipe = Pipeline([
    ('lr', lr)
])

params = {
    'lr__penalty':['l1','l2'],
    'lr__C':[0.65,0.7,0.75],
    'lr__class_weight':['balanced',None],
    'lr__max_iter':[15,20]
}
gs = GridSearchCV(pipe, param_grid=params)
gs.fit(X_train_nf, y_train.values.ravel())
print(gs.best_score_)
gs.best_params_
```

```python
logreg = LogisticRegression(C=.65, 
                            penalty='l1', 
                            class_weight='balanced',
                            max_iter=15
                           )
logreg.fit(X_train_sc_nf, y_train.values.ravel())
logreg.score(X_test_sc_nf, y_test)
```

### We can see that this score with the extended features, was an improvement over the score with the original data.  Of course, accuracy is not what we are interested.

### The single biggest difference in tuning came from the choice of regularizer.  Convergence took no time at all.

```python
predictions_logreg = logreg.predict(X_test_sc_nf)
```

```python
confusion_matrix(y_test, predictions_logreg)
```

### The True-Positive Rate (TPR) for this model is ...

```python
1021 / (1021+637)
```

### ...61.6%.  Let's check the f1-score which is a balance between precision and recall.

```python
print(classification_report(y_test, predictions_logreg))
```

```python
X_test_copy_lr = X_test.copy()
X_test_copy_lr['y_hat'] = predictions_logreg
```

```python
logreg_cm = pd.DataFrame(confusion_matrix(y_test, X_test_copy_lr['y_hat']), 
                         columns=["","True Oct-Default"])
```

## E. CaRT Modeling:

```python
dt = DecisionTreeClassifier()

pipe = Pipeline([
    ('dt', dt)
])

params = {
    'dt__criterion':['gini','entropy'],
    'dt__class_weight':['balanced',None],
    'dt__max_depth':[6,8,12,None],
    'dt__min_samples_split':[2,3,4],
    'dt__min_samples_leaf':[1,2]
}
gs = GridSearchCV(pipe, param_grid=params)
gs.fit(X_train_nf, y_train)
print(gs.best_score_)
gs.best_params_
```

### There is no choice for us to use "class_weight = 'balanced'," unfortunately.

```python
cart_class = DecisionTreeClassifier(criterion='gini',
                                    class_weight = 'balanced',
                                    max_depth=6,
                                    min_samples_split=4, 
                                    min_samples_leaf=1)
model_cart = cart_class.fit(X_train_nf, y_train)
feature_columns = X_train_nf.columns
X_train_nf_copy=X_train_nf.copy()
X_test_nf_copy=X_test_nf.copy()
X_train_nf_copy['y_hat'] = model_cart.predict(X_train_nf)
X_test_nf_copy['y_hat'] = model_cart.predict(X_test_nf)
```

```python
f1_train_nf_cart = f1_score(y_train,
                    X_train_nf_copy['y_hat'])
f1_test_nf_cart = f1_score(y_test,
                   X_test_nf_copy['y_hat'])
```

```python
print("The training F1-score for " + str(cart_class) + " is: " + str(f1_train_nf_cart))
print("The testing F1-score for " + str(cart_class) + " is: " + str(f1_test_nf_cart))
```
